```python
wine_reviews_points_means=wine_reviews.points.mean()
```

```python
wine_reviews.points.map(lambda p: p - wine_reviews_points_means)
```

```python
def remean_points(row):
    row.points = row.points - wine_reviews_points_means
    return row

wine_reviews.apply(remean_points, axis='columns')
```
